```python
%matplotlib widget
%reload_ext autoreload
%autoreload 2
%matplotlib notebook
import matplotlib.pyplot as plt

from Utilities.downloaddata import fetch_data as fdata
import SimpleITK as sitk

nifti_path='Data/training_001_mr_T1.nii'
mha_img = sitk.ReadImage(fdata('training_001_mr_T1.mha'))
sitk.WriteImage(mha_img, nifti_path)
```

```python
#Let's make sure if it is converted correctly
npa = sitk.GetArrayViewFromImage(mha_img)[10,:,:]
fig = plt.figure(figsize=(10,3))
fig.add_subplot(1,2,1)
plt.imshow(npa,cmap=plt.cm.Greys_r);
plt.title('grey colormap-mda', fontsize=10)
plt.axis('off')

nifti_img = sitk.ReadImage('Data/training_001_mr_T1.nii')
npa = sitk.GetArrayViewFromImage(nifti_img)[10,:,:]
fig.add_subplot(1,2,2)
plt.imshow(npa,cmap=plt.cm.Greys_r);
plt.title('grey colormap-nifti', fontsize=10)
plt.axis('off')

```
